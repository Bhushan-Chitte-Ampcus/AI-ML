```python
# For chat conversation

from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2:1b",
    temperature=0
)

response = llm.invoke("Explain the concept of quantum entanglement in one sentence")
response.content
```

```python
# For non-chat execution

from langchain_ollama import OllamaLLM

llm = OllamaLLM(
    model="llama3.2:1b"
)

response = llm.invoke("The capital of India is")
response
```

```python
# Ollama embeddings

from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model="embeddinggemma:latest"
)

query_result = embeddings.embed_query("What is langchain?")
print(query_result)
```

```python
# Embed multiple documents

doc_results = embeddings.embed_documents(["doc1", "doc2"])
print(doc_results)
```
