```python
import ollama
import base64
```

```python
response = ollama.generate(model="qwen2.5:7b", prompt="Why does moon glow?")
print(response)
```

```python
response = ollama.generate(model="qwen2.5:7b", prompt="Why does moon glow?", stream=True)

for chunk in response:
    print(chunk["response"], end="")
```

```python
img_path = "./img2.jpg"

with open(img_path, "rb") as f:
    image_bytes = f.read()

image_64 = base64.b64encode(image_bytes).decode("utf-8")

response = ollama.generate(
    model="gemma3:4b",
    images=[image_64],
    prompt="Give a caption for this image."
)

print(response.response)
```

```python
images_path = ["./img1.png", "./img2.jpg"]

images_base64 = []
for i in images_path:
    with open(i, "rb") as f:
        images_bytes = f.read()
        images_base64.append(base64.b64encode(images_bytes).decode("utf-8"))

response = ollama.generate(
    model="gemma3:4b",
    images=images_base64,
    prompt="Generate a story based on these images, make sure you take context from each and every image."
)

print(response.response)
```

```python
# SYSTEM INSTRUCTIONS

response = ollama.generate(
    model="qwen2.5:7b",
    prompt="Why does moon glow?",
    system="You are a funny assistant, you explain things in funny ways."
)

print(response.response)
```

```python
# OPTIONS PARAMETER

response = ollama.generate(
    model="qwen2.5:7b",
    prompt="Why does moon glow?",
    options={
        "temperature": 0.3,
        "top_p": 0.5,
        "top_k": 45
    }
)

print(response.response)
```

```python
# OLLAMA LIST

local_models = ollama.list()
print(local_models)

print("*"*80)

for model in local_models["models"]:
    print(model["model"], model["size"])
```

```python
# OLLAMA PULL

model_name = "llama3.2:1b"
progress = ollama.pull(model_name, stream=True)

for i in progress:
    print(i)
```
