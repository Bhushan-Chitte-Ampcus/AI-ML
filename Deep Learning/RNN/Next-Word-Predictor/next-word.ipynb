```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Bidirectional, Dropout
from keras.callbacks import EarlyStopping
```

```python
def clean_text(text):
    text = text.replace('\r\n', '\n').replace('\r', '\n')

    cleaned_lines = []

    for line in text.split('\n'):

        line = re.sub(r'#+', '', line)
        line = re.sub(r'\*\*', '', line)
        line = re.sub(r'=+|-+', ' ', line)

        # keep numbers but remove unwanted symbols
        line = re.sub(r'[^a-zA-Z0-9.,()\-\s]', ' ', line)

        line = re.sub(r'[ \t]+', ' ', line).strip().lower()

        if line:
            line = "<start> " + line + " <end>"
            cleaned_lines.append(line)

    return "\n".join(cleaned_lines)
```

```python
cleaned_text = clean_text(text)
print(cleaned_text)
```

```python
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts([cleaned_text])
```

```python
tokenizer.word_index
```

```python
input_sequences = []

for sentence in cleaned_text.split("\n"):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)):
        n_gram = tokenized_sentence[:i+1]
        input_sequences.append(n_gram)
```

```python
input_sequences
```

```python
max_len = max([len(i) for i in input_sequences])
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding="pre")
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
y = to_categorical(y, num_classes=len(tokenizer.word_index) + 1)
y
```

```python
y.shape
```

- input_dim → Vocabulary size
- output_dim → Embedding vector size
- input_length → Sequence length

```python
model = Sequential()

model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=128, input_length=X.shape[1]))

model.add(Bidirectional(LSTM(150, return_sequences=True)))
model.add(Dropout(0.2))

model.add(LSTM(150))
model.add(Dropout(0.2))

model.add(Dense(len(tokenizer.word_index) + 1, activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
early_stop = EarlyStopping(
    monitor="loss",
    patience=5,
    restore_best_weights=True
)
```

```python
history = model.fit(X, y, epochs=100, batch_size=64, callbacks=[early_stop])
```

```python
model.fit(X, y, epochs=200, initial_epoch=99, batch_size=64, callbacks=[early_stop])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,6))

sns.lineplot(x = history.epoch, y = history.history["loss"], ax=ax[0], label="training loss")
sns.lineplot(x = history.epoch, y = history.history["accuracy"], ax=ax[1], label="training accuracy")

plt.tight_layout()
plt.show()
```

### Prediction

```python
# create reverse index once
index_word = {index: word for word, index in tokenizer.word_index.items()}

def sample_with_temperature(preds, temperature=0.8):
    preds = np.asarray(preds).astype("float64")
    
    preds = np.log(preds + 1e-10) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)

    return np.random.choice(len(preds), p=preds)


def generate_text(model, seed_text, next_words=10, temperature=0.8):

    for _ in range(next_words):

        # convert text to tokens
        token_list = tokenizer.texts_to_sequences([seed_text])[0]

        # pad sequence
        token_list = pad_sequences([token_list], maxlen=max_len, padding='pre')

        # predict probabilities
        preds = model.predict(token_list, verbose=0)[0]

        # sample word index
        predicted_index = sample_with_temperature(preds, temperature)

        # get word from index
        output_word = index_word.get(predicted_index, "")

        if output_word == "":
            break

        seed_text += " " + output_word

    return seed_text
```
