### Sentiment Analysis - (Integer Encoding)

```python
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data()
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
X_train = pad_sequences(X_train, padding="post", maxlen=400)
X_test = pad_sequences(X_test, padding="post", maxlen=400)
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(1000,1), return_sequences=False))
model.add(Dense(1, activation="sigmoid"))

model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

### IMDB Dataset

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
```

```python
X_train = pad_sequences(X_train, maxlen=400, padding="post")
X_test = pad_sequences(X_test, maxlen=400, padding="post")
```

```python
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=32))
model.add(SimpleRNN(32))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))
```

```python
def predict_sentiment(text, model, tokenizer, max_len=400):
    """
    Predict sentiment of a given text using a trained SimpleRNN model.
    
    Args:
        text (str): Input sentence
        model: Trained Keras model
        tokenizer: Fitted tokenizer
        max_len (int): Max sequence length used during training
    
    Returns:
        dict: sentiment label and probability
    """
    
    # Convert text to sequence
    sequence = tokenizer.texts_to_sequences([text])
    
    # Pad sequence
    padded = pad_sequences(sequence, maxlen=max_len, padding='post', truncating='post')
    
    # Predict
    prediction = model.predict(padded)[0][0]
    
    # Convert to label
    if prediction >= 0.5:
        sentiment = "Positive"
    else:
        sentiment = "Negative"
    
    return {
        "text": text,
        "sentiment": sentiment,
        "confidence": float(prediction)
    }
```

```python
text = "I really loved this movie!"
result = predict_sentiment(text, model, tokenizer)

print(result)
```
